```python
# Import required libraries
from pathlib import Path
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Create working copies of DataFrames
customers_df = pd.DataFrame(customers_data)
orders_df = pd.DataFrame(orders_data)
payments_df = pd.DataFrame(payments_data)
```

```python
# Convert order date columns to Datetime format
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
orders_df[date_cols] = orders_df[date_cols].apply(lambda cols: pd.to_datetime(cols,errors='coerce'))
```

```python
# Check missing approved data by order status
orders_df.groupby('order_status')['order_approved_at'].apply(lambda x: x.isna().sum())
```

```python
# Fill missing approved Dates for delivered orders
# Using Purchase Timestamp
orders_df.loc[(orders_df['order_approved_at'].isna()) & 
    (orders_df['order_status'] == 'delivered'),'order_approved_at'] = orders_df['order_purchase_timestamp']  
```

```python
# Check missing values carrier delivery dates by order status
orders_df.groupby('order_status')['order_delivered_carrier_date'].apply(lambda x: x.isna().sum())
```

```python
# Fill missing carrier delivery dates
# Using approved date delivered & shipped orders
mask_carrier = (
    orders_df['order_delivered_carrier_date'].isna() &
    orders_df['order_status'].isin(['delivered','shipped'])
)
orders_df.loc[mask_carrier,'order_delivered_carrier_date'] = orders_df.loc[mask_carrier,'order_approved_at']
```

```python
# Check using customer delivery dates by orders status
orders_df.groupby('order_status')['order_delivered_customer_date'].apply(lambda x: x.isna().sum())
```

```python
# Fill missing customer delivery dates 
# Using estimated delivery date for delivered orders
mask_customer = (
    orders_df['order_delivered_customer_date'].isna() &
    (orders_df['order_status'] == 'delivered')
)
orders_df.loc[mask_customer,'order_delivered_customer_date'] = orders_df.loc[mask_customer,'order_estimated_delivery_date']
```

```python
# Create MySQL connection
con = create_engine('mysql+pymysql://root:12345@localhost/olist_commerce_analytics_db')

# Export cleaned data to MySQL
customers_df.to_sql(
    'olist_customers_data',
    con = con,
    if_exists = 'replace',
    index = False
)

orders_df.to_sql(
    'olist_orders_data',
    con = con,
    if_exists = 'replace',
    index = False
)

payments_df.to_sql(
    'olist_payments_data',
    con = con,
    if_exists = 'replace',
    index = False
)

# Success message
print('Data Successfully Imported✅')
```
